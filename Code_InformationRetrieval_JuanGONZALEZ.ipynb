### 6. Invert index matrix

```python
from collections import defaultdict
def create_index (data):
    index = defaultdict(list)
    for i, tokens in enumerate(data):
        for token in tokens:
            index[token].append(i+1)
    return index
```

```python
index = create_index([new_term_vector1, new_term_vector2,new_term_vector3,new_term_vector4,new_term_vector5,new_term_vector6,new_term_vector7,new_term_vector8,new_term_vector9,new_term_vector10])
```
